```python
'''
 Based on the following tutorial:
   http://docs.opencv.org/3.0-beta/doc/py_tutorials/py_feature2d/py_fast/py_fast.html
'''

import numpy as np
import cv2
import time

# Load the image
img = cv2.imread('croatiadark.jpg')

# Convert it to gray scale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Detect the FAST key points

ctime = 0.000


for i in range(100):
    #set Start time 
    st = time.time()

    fast = cv2.FastFeatureDetector_create(threshold=50)
    keyPoints = fast.detect(gray, None)


    # Paint the key points over the original image
    result = cv2.drawKeypoints(img, keyPoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

    # wait for 3 seconds
    time.sleep(3)

    #get end time
    et = time.time()
    # get the execution time
    elapsed_time = et - st
    ctime= ctime + float(elapsed_time)
    print("Loading" ,i + 1,"%")

print("Execution time:",ctime /100,"seconds")
print("Total Keypoints with nonmaxSuppression: ", len(keyPoints))
# Display the results
cv2.imshow('Key points', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2

# Load the image
img = cv2.imread('croatiadark.jpg')

# Convert it to gray scale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


ctime = 0.000


for i in range(100):
    #set Start time 
    st = time.time()

    # Detect the Harris corners
    gray = np.float32(gray)
    cornersImg = cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)

    # Dilate the image to increase the corners size
    cornersImg = cv2.dilate(cornersImg, np.ones((3, 3)))

    # Paint the corners over the original image
    img[cornersImg > 0.2 * cornersImg.max()] = [0, 0, 255]

    # wait for 3 seconds
    time.sleep(3)

    #get end time
    et = time.time()
    # get the execution time
    elapsed_time = et - st
    ctime= ctime + float(elapsed_time)
    print("Loading" ,i + 1,"%")

print("Execution time:",ctime/100,"seconds")

# Display the results
cv2.imshow('Corners', img)
cv2.waitKey(0)
cv2.destroyAllWindows()

```
